```python
import json
```

```python
# Let's write a function to load the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

```python
def display_users(data):
    print("Users and ther connections: \n")
    for user in data["users"]:
        print(f"ID: {user['id']} - {user['name']}'s friends are {user['friends']} and they liked this pages {user['liked_pages']}")
    print("\nPages informations: \n")
    for page in data["pages"]:
        print(f"{page['id']}: {page["name"]}")

display_users(data)
```
